```python
import torch
import torchvision.models as models
import torchvision.transforms as transforms
```

```python
model = models.resnet18(pretrained = True)
```

```python
transform = transforms.Compose([
    transforms.RandomRotation(20),
    transforms.RandomHorizontalFlip(0.5),
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])
```

```python
from torchvision.datasets import ImageFolder
import shutil
import os

#Load data
data = ImageFolder(root="train_images",transform=transform)
print(data)
```

```python
print(data.classes)
print(data.targets)
print(len(data.targets))
```

```python
from collections import Counter
print(Counter(data.targets))
```

```python
train_size = int(0.8 * len(data))
test_size = len(data) - train_size
train_data, test_data = torch.utils.data.random_split(data, [train_size,test_size])

train_targets = [data.targets[i] for i in train_data.indices]
test_targets = [data.targets[i] for i in test_data.indices]
```

```python
print(train_targets)
print(test_targets)
```

```python
from torch.utils.data import DataLoader

batch_size = 8

train_dataloader = DataLoader(train_data, batch_size=batch_size)
test_dataloader = DataLoader(test_data, batch_size=batch_size)
```

```python
for X in train_dataloader:
    print(f"shape of X[N,C,H,W]: {X[0].shape}")
```

```python
from torch import nn

number_features = model.fc.in_features
# print(number_features)
model.fc = nn.Linear(number_features, 3)

for param in model.parameters():
    param.requires_grad = False

for param in model.fc.parameters():
    param.requires_grad = True
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)
```

```python
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()

    for batch, (X,y) in enumerate(dataloader):
        pred = model(X)
        loss = loss_fn(pred, y)

        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        if batch % 2 == 0:
            loss = loss.item()
            current_input = batch * len(X)
            print(f"loss: {loss:f} [{current_input:>5d}/{size:>5d}")
```

```python
def test(dataloader, model, loss_fs):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()

    test_loss = 0
    correct = 0

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)

            test_loss = test_loss + loss_fn(pred, y).item()

            correct_numbers_in_batch = (pred.argmax(1) == y).type(torch.float).sum().item()
            correct = correct + correct_numbers_in_batch
        test_loss = test_loss / num_batches
        correct = correct / size

        print(f"Test Error: \n Accuracy: (correct / # of images) {100*correct:>0.1f}%, Average Loss: {test_loss:>8f} \n")
```

```python
epochs = 15
for t in range(epochs):
    print(f"Epoch {t+1}\n ================")
    train(train_dataloader, model, loss_fn, optimizer)
    test(test_dataloader, model, loss_fn)

print("Training is Finished!")
```
